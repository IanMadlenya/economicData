```python
from __future__ import division
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as dts
import pandas as pd
import wbdata
import runProcs
from scipy import stats
import qtyTheoryFunc as qt

# if wbdata fails to load run in the shell: rm -r ~/.cache/wbdata
```

```python
# 0. Preliminaries

# 0.1 general plot settings

font = {'weight' : 'bold',
        'size'   : 15}
axes={'labelweight' : 'bold'}
plt.rc('font', **font)
plt.rc('axes', **axes)
plt.rcParams['xtick.major.pad']='8'
plt.rcParams['ytick.major.pad']='8'
```

```python
# 1. Get the World Bank ISO codes
isoCodes = pd.DataFrame(wbdata.search_countries("",display=False))
isoCodes = isoCodes[['id','name']]
isoCodes = isoCodes.set_index('name')
```

```python
# 2. Indices for data series and labels for data frame

indicators = {'FP.CPI.TOTL': 'inflation rate',
              'FM.LBL.MONY.CN': 'money growth', 
              'NY.GDP.MKTP.KD': 'real gdp growth'
              }
```

```python
# 3. Create the data frames with raw money, price level, and real gdp data from WB

# 3.1 country groups
countries =  [i['id'] for i in wbdata.get_country(incomelevel=['LIC','MIC','HIC'],display=False)]
countriesH = [i['id'] for i in wbdata.get_country(incomelevel=['HIC'],display=False)]
countriesM = [i['id'] for i in wbdata.get_country(incomelevel=['MIC'],display=False)]
countriesL = [i['id'] for i in wbdata.get_country(incomelevel=['LIC'],display=False)]

# 3.2 data frames
df   = wbdata.get_dataframe(indicators, country=countries,convert_date=True)
hicDf = wbdata.get_dataframe(indicators, country=countriesH,convert_date=True)
micDf = wbdata.get_dataframe(indicators, country=countriesM,convert_date=True)
licDf = wbdata.get_dataframe(indicators, country=countriesL,convert_date=True)
```

```python
# 4. Create data sets of money growth, inflation, and real gdp growth
qtyTheoryData, include, dropped = qt.create(dataFrame=df,indicators=indicators,levels=[''],csvFilename='qtyTheoryData',decimals=5)
qtyTheoryDataH, includeH, droppedH = qt.create(dataFrame=hicDf,indicators=indicators,levels=[''],csvFilename='qtyTheoryDataH',decimals=5)
qtyTheoryDataM, includeM, droppedM = qt.create(dataFrame=micDf,indicators=indicators,levels=[''],csvFilename='qtyTheoryDataM',decimals=5)
qtyTheoryDataL, includeL, droppedL = qt.create(dataFrame=licDf,indicators=indicators,levels=[''],csvFilename='qtyTheoryDataL',decimals=5)
```

```python
# 6. Verify output 
print 'Number of countries:',len(include),'\n'
print qtyTheoryData.loc['United States']
print qtyTheoryData.loc['Germany']
```

```python
# 7. Plots.

# 7.1 Money growth and inflation

xmin = -0.05
xmax = 1.4
ymin = -0.05
ymax = 1.4

x45 = np.arange(xmin,xmax,0.001)
y45 = x45 - np.mean(qtyTheoryData['real gdp growth'])

fig = plt.figure()

ax1 = fig.add_subplot(1, 1, 1)
ax1.plot(x45,y45)
ax1.plot(qtyTheoryDataH['money growth'],qtyTheoryDataH['inflation rate'],'bo')
ax1.plot(qtyTheoryDataM['money growth'],qtyTheoryDataM['inflation rate'],'gs')
ax1.plot(qtyTheoryDataL['money growth'],qtyTheoryDataL['inflation rate'],'r^')
plt.grid(True)
# ax1.set_title('Money growth and inflation',fontsize=15)
ax1.set_xlabel('money growth')
ax1.set_ylabel('inflation')
plt.legend(['$45^\circ$'],loc='lower right',fontsize='15')
plt.axis([xmin,xmax,ymin,ymax])

plt.tight_layout()
plt.savefig('fig_moneyInflation.png',bbox_inches='tight',dpi=120)
```

```python
# 7.2 Money growth and real gdp growth

xmin = -0.05
xmax = 1.4
ymin = -0.05
ymax = .2

x0 = np.arange(xmin,xmax,0.001)
y0 = x0*0 + np.mean(qtyTheoryData['real gdp growth'])

fig = plt.figure()

ax1 = fig.add_subplot(1, 1, 1)
ax1.plot(x0,y0)
plt.plot(qtyTheoryDataH['money growth'],qtyTheoryDataH['real gdp growth'],'bo')
plt.plot(qtyTheoryDataM['money growth'],qtyTheoryDataM['real gdp growth'],'gs')
plt.plot(qtyTheoryDataL['money growth'],qtyTheoryDataL['real gdp growth'],'r^')
plt.grid(True)
# ax1.set_title('Money growth and real GDP growth',fontsize=15)
ax1.set_xlabel('money growth')
ax1.set_ylabel('real GDP growth')
plt.axis([xmin,xmax,ymin,ymax])
plt.legend(['$0^\circ$'],loc='lower right',fontsize='15')

plt.tight_layout()
plt.savefig('fig_moneyRealGDP.png',bbox_inches='tight',dpi=120)
```

```python
# 7.3 Money growth and real gdp growth

xmin = -0.05
xmax = 1.4
ymin = -0.05
ymax = .2

x0 = np.arange(xmin,xmax,0.001)
y0 = x0*0 + np.mean(qtyTheoryData['real gdp growth'])

fig = plt.figure()

ax1 = fig.add_subplot(1, 1, 1)
ax1.plot(x0,y0)
plt.plot(qtyTheoryDataH['inflation rate'],qtyTheoryDataH['real gdp growth'],'bo')
plt.plot(qtyTheoryDataM['inflation rate'],qtyTheoryDataM['real gdp growth'],'gs')
plt.plot(qtyTheoryDataL['inflation rate'],qtyTheoryDataL['real gdp growth'],'r^')
plt.grid(True)
# ax1.set_title('Inflation and real GDP growth',fontsize=15)
ax1.set_xlabel('inflation')
ax1.set_ylabel('real GDP growth')
plt.axis([xmin,xmax,ymin,ymax])
plt.legend(['$0^\circ$'],loc='lower right',fontsize='15')

plt.tight_layout()
plt.savefig('fig_inflationRealGDP.png',bbox_inches='tight',dpi=120)
```

```python
# 10 Money growth and inflation with ISO codes as labels and an OLS regression line

m = qtyTheoryData['money growth']
p = qtyTheoryData['inflation rate']
y = qtyTheoryData['real gdp growth']
codes = qtyTheoryData['iso code']


a_p, b_p, r_p, p_value, std_err = stats.linregress(m,p)
xVals = np.arange(-10,10)
p_pred = b_p + a_p*xVals

fig = plt.figure()

ax1 = fig.add_subplot(1, 1, 1)
plt.plot(p_pred,p_pred,lw=1.5,label='OLS Regression')
for i in range(len(codes)):
    if codes[i] =='USA':
        usaIndex = i
#         plt.text(m[i], p[i], codes[i], color='red',fontsize=12, clip_on=True,horizontalalignment='center',verticalalignment='center',alpha = 1)
    else:
        plt.text(m[i], p[i], codes[i], color="#11557c",fontsize=12, clip_on=True,horizontalalignment='center',verticalalignment='center',alpha = 1)
plt.text(m[usaIndex], p[usaIndex], codes[usaIndex], color='red',fontsize=12, clip_on=True,horizontalalignment='center',verticalalignment='center',alpha = 1)
plt.axis([-0.05, 1.4, -0.05, 1.4])
print 'regression coefficient:',np.round(a_p,2)
```
